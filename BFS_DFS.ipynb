```python
def read_graph_from_file(filename):
    matrix = []
    try:
      with open(filename, 'r') as file:
        for line in file:
            # Tách dòng thành các sổ, chuyển sang kiểu int và tạo thành một hàng của ma trận
            row = [int(val) for val in line.strip().split()]
            matrix.append(row)
    except FileNotFoundError:
          print(f"Loi: Khong tim thay file '{filename}'")
    return matrix
print("matrankedoctufile")
matrix = read_graph_from_file("graph_exampl.txt")
print(matrix)
```

```python
import matplotlib.pyplot as plt
class Graph:
  "Mot class bieu dien do thi tu ma tran lien ke adj-matrix"
  "Bao gom cac phuong thuc de hien thi, ve va duyet do thi"
  def __init__(self, adj_matrix):
    self.adj_matrix = adj_matrix
    self.num_vertices = len(adj_matrix)
  def __str__(self):
    "tra ve chuoi bieu dien ma tran ke de in ra cho dung"
    if not self.adj_matrix:
      return "Do thi rong"
    matrix_str = ""
    for row in self.adj_matrix:
      matrix_str += " ".join(map(str,row)) + "\n"
    return matrix_str
  def bfs(self, start_vertex):
    "duyet do thi theo chieu rong tu dinh bat dau"
    if start_vertex < 0 or start_vertex >= self.num_vertices:
      print(f"Loi: dinh bat dau {start_vertex} khong hop le")
      return
    print(f"Duyet do thi theo chieu rong tu dinh {start_vertex}:")
    queue = [start_vertex]
    visited = {start_vertex}
    while queue:
      current_vertex = queue.pop(0)
      print(current_vertex, end=" ")
      for neighbor in range(self.num_vertices):
        if self.adj_matrix[current_vertex][neighbor] == 1 and neighbor not in visited:
          queue.append(neighbor)
          visited.add(neighbor)
          print(f" -> Di chuyen qua canh {current_vertex}-{neighbor}")
          print(f"Tham dinh {neighbor}")
    print(f"\n--------Ket Thuc Ham BFS tai dinh {start_vertex}-------")
  def draw(self, cmap_name='gray'): # Thêm tham số cmap_name tùy chọn với giá trị mặc định là 'gray'
    "ve do thi bang cach dung thu vien matplotlib"
    if not self.adj_matrix:
      print("Do thi rong")
      return
    plt.figure(figsize=(5,5))
    plt.imshow(self.adj_matrix, cmap=cmap_name, interpolation='nearest') # Sử dụng cmap_name ở đây
    plt.colorbar()
    plt.show()
Graph = Graph(matrix)
print(Graph)
for i in range (0,4) :
  Graph.bfs(i)
```
